```python
def test(bsnsCond):
  URL = ('http://apis.data.go.kr/6260000/BusanTblFnrstrnStusService/getTblFnrstrnStusInfo?serviceKey=zpYTrUIjrmxMwfmBZ8ZpleWRntXuuWi%2Bq%2BWr%2BSn1VHO7OOT3NTV86KHTsvtFy7WtoCbnrAExXmJWeGcJOy0E7A%3D%3D&numOfRows=10&pageNo=1&resultType=json')
  
  response = requests.get(URL)
  r = response.json()
  items = r['getTblFnrstrnStusInfo']['item']

  addrJibun = [item['addrJibun'] for item in items]
  addrRoad = [item['addrRoad'] for item in items]
  bsnsCond = [item['bsnsCond'] for item in items]
  bsnsNm = [item['bsnsNm'] for item in items]
  menu = [item['menu'] for item in items]
  tel = [item['tel'] for item in items]

  addrJibun = pd.Series(addrJibun)
  addrRoad = pd.Series(addrRoad)   
  bsnsCond = pd.Series(bsnsCond)
  bsnsNm = pd.Series(bsnsNm)
  menu = pd.Series(menu)
  tel = pd.Series(tel)


  df = pd.DataFrame({'addrJibun': addrJibun, 'addrRoad': addrRoad, 'bsnsCond': bsnsCond, 'bsnsNm': bsnsNm, 'menu': menu, 'tel': tel})
  return df
```
